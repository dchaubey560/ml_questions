```python
import pandas as pd
from sklearn.metrics import accuracy_score,r2_score,confusion_matrix,recall_score,precision_score,mean_absolute_error
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
data=load_breast_cancer()
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df["target"]=data.target
df
```

```python
X=df.drop("target",axis=1)
y=df.iloc[:,-1]
X
y
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier()
parameters={
    "n_estimators":[100,150,200,250,300],
    "max_depth":[3,4,5,6,7]
}
grid=GridSearchCV(model,parameters,cv=5)
grid.fit(Xtrain,ytrain)
```

```python
print("Training Efficiency :",grid.score(Xtrain,ytrain))
print("Training Efficiency :",grid.score(Xtest,ytest))
print(grid.best_params_)
```
