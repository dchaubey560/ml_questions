```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer,fetch_california_housing
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
```

```python
lbc=load_breast_cancer()
df=pd.DataFrame(lbc.data,columns=lbc.feature_names)
df
df["cancer"]=lbc.target
df.head()
```

```python
X=df.drop("cancer",axis=1)
y=df.cancer
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=150,max_depth=5)
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
ypred
```

```python
print("Training Efficiency :",model.score(Xtrain,ytrain))
print("Training Efficiency :",model.score(Xtest,ytest))
```

```python
print('Accuracy score :',accuracy_score(ytest,ypred))
print('Precision score',precision_score(ytest,ypred,average="weighted"))
print('Recall Score',recall_score(ytest,ypred,average="weighted"))
```
