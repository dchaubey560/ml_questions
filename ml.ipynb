```python
df=pd.read_csv('https://github.com/Jeevita-R/ML_datasets/raw/refs/heads/main/House_price_prediction.csv')
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True,dtype=int)
df
```

```python
df=df.drop("Unnamed: 0",axis=1)
df
```

```python
#divide the data
X=df.drop ('Price',axis=1)
y=df['Price']
X
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42) #random state if not provide data shuffle between testing and training 
```

```python
Lr=LinearRegression()
Lr.fit(Xtrain,ytrain)
LinearRegression()
```

```python
ypred=Lr.predict(Xtest)
ypred
```

```python
print("Training",Lr.score(Xtrain,ytrain))
print("Testing",Lr.score(Xtest,ytest))
```

```python
print('MAE',mean_absolute_error(ytest,ypred))
```

```python
#RIEDGE REGRESSION
ridge=Ridge(alpha=1)
ridge.fit(Xtrain,ytrain)
```

```python
ypred=ridge.predict(Xtest)
ypred
```

```python
print("Training Err",ridge.score(Xtrain,ytrain))
print("Training Err",ridge.score(Xtest,ytest))
```

```python
lasso=Lasso(alpha=5)
lasso.fit(Xtrain,ytrain)
print(lasso)
#Lasso(alpha=55)
```

```python
ypred=lasso.predict(Xtest)
ypred
```

```python
print("Training Eff",lasso.score(Xtrain,ytrain))
print("Training Eff",lasso.score(Xtest,ytest))
```
