```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.datasets import load_breast_cancer
```

```python
df = pd.read_excel(r"C:\Users\ASUS\Downloads\house_price_regression_dataset.csv")
```

```python
#divide the data into input and output feature
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=GradientBoostingRegressor(n_estimators=300)
model.fit(Xtrain,ytrain)
```

```python
print("Training Efficiency :",model.score(Xtrain,ytrain))
print("Training Efficiency :",model.score(Xtest,ytest))
```

```python
ypred=model.predict(Xtest)
```

```python
print("R2 ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

```python
#classifiers
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,recall_score,f1_score,confusion_matrix,accuracy_score
```

```python
df = pd.read_excel(r"C:\Users\ASUS\Downloads\bankloan.csv")
df
```

```python
df = df.drop(['ID', 'ZIP.Code'], axis=1)
df
```

```python

#keep personal loan as y
X=df.drop("Personal.Loan",axis=1)
y=df["Personal.Loan"]
y
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=GradientBoostingClassifier(n_estimators=300,learning_rate=0.02)
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
ypred
```

```python
print("Training Efficiency :",model.score(Xtrain,ytrain))
print("Testing Efficiency :",model.score(Xtest,ytest))
```

```python
confusion_matrix(ytest,ypred)
```

```python
print('Precision score',precision_score(ytest,ypred))
print('Recall Score',recall_score(ytest,ypred))
print('Accuracy Score',accuracy_score(ytest,ypred))
```
