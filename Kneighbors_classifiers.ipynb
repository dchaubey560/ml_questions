```python
import pandas as pd
import numpy as np#by default average set to be binary,when we have multi class data in output columns different then we use weighted
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_excel(r"C:\Users\ASUS\Downloads\da by sourav sir and ml by qspidermam\Iris.xlsx")
data
```

```python
X=data.iloc[:,:4]
X
StandardScaler().fit_transform(X)
```

```python
y=data.iloc[:,-1]
y
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
Knc=KNeighborsClassifier(n_neighbors=5)
Knc.fit(Xtrain,ytrain)
```

```python
KNeighborsClassifier()
```

```python
ypredc=Knc.predict(Xtest)
ypredc
```

```python
print("Training Efficiency :",Knc.score(Xtrain,ytrain))
print("Training Efficiency :",Knc.score(Xtest,ytest))
```

```python
print('Accuracy score :',accuracy_score(ytest,ypredc))
print('Precision score',precision_score(ytest,ypredc,average="weighted"))
print('Recall Score',recall_score(ytest,ypredc,average="weighted"))
```
