```python
import pandas as pd
import numpy as np#by default average set to be binary,when we have multi class data in output columns different then we use weighted
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
                #Taxi trip pricing for regression if cleaning is required then 
df=pd.read_csv(r"C:\Users\ASUS\Downloads\taxi_trip_pricing (1).csv")
df.head()
```

```python
df = df.dropna()
df=df.drop(['Weather', 'Passenger_Count', 'Traffic_Conditions','Day_of_Week'], axis=1)
df
```

```python
X = df.drop('Trip_Price',axis=1)
y = df['Trip_Price']
```

```python
X=pd.get_dummies(df,dtype=int,drop_first=True)
X.head()
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
Knr=KNeighborsRegressor(n_neighbors=5)
Knr.fit(Xtrain,ytrain)
```

```python
KNeighborsRegressor()
```

```python
ypredc=Knr.predict(Xtest)
ypredc
```

```python
print("Training Efficiency :",Knr.score(Xtrain,ytrain))
print("Training Efficiency :",Knr.score(Xtest,ytest))
```

```python
print("MAE :", mean_absolute_error(ytest, ypredc))
print("MSE :", mean_squared_error(ytest, ypredc))
print("RMSE:", np.sqrt(mean_squared_error(ytest, ypredc)))
print("R2 Score :", r2_score(ytest, ypredc))
```
